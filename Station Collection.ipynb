```python
def get_this_county_state_station(driver, county, state):
    state = state_dict[state]
    
    driver.find_element_by_tag_name('option').click()
    page_options = driver.find_elements_by_tag_name('option')
    for opt in page_options:
#         print(opt.text)
        if opt.text == 'County':
            opt.click()
            break
    sleep(0.5)
    page_options = driver.find_elements_by_tag_name('option')
    for opt in page_options:
#         print(opt.text)
        if opt.text == state:
            opt.click()
            break
    sleep(0.5)
    page_options = driver.find_elements_by_tag_name('option')
    for opt in page_options:
#         print(opt.text)
        if county in opt.text:
            opt.click()
            break
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "stationID"))
    )
    page_stations = driver.find_elements_by_class_name('stationID')
    max_station = '0'
    max_station_name = ''

    for station in page_stations:
        for i, char in enumerate(station.text[::-1]):
            if not char.isdigit():
#                 print(station.text[-i:])
                if station.text[-i:] > max_station and not station.text.startswith('US1'):
                    max_station = station.text[-i:]
                    max_station_name = station.text
                break

    return (max_station_name[max_station_name.find(':')+1:])
```
